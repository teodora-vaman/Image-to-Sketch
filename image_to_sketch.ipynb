```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage import io, color
import scipy.ndimage.morphology as morpho
```

```python
image_path = "images\lena.jpg"
img = cv2.imread(image_path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

grayscale_image = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
inverted_image = cv2.bitwise_not(grayscale_image)
smooth_image = cv2.GaussianBlur(grayscale_image, (5,5), 0);
sketch_image1 = cv2.divide(grayscale_image, smooth_image, scale=256.0)
smooth_image = cv2.GaussianBlur(inverted_image, (5,5), 0)
sketch_image2 = cv2.divide(grayscale_image, 255 - smooth_image, scale=256.0)


plt.figure(figsize=(20,20))
plt.subplot(1,5,1)
plt.imshow(img, cmap='gray')
plt.title("Original Image")
plt.subplot(1,5,2)
plt.imshow(grayscale_image, cmap='gray')
plt.title("Gray Image")
plt.subplot(1,5,3)
plt.imshow(inverted_image, cmap='gray')
plt.title("Inverted Image")
plt.subplot(1,5,4)
plt.imshow(smooth_image, cmap='gray')
plt.title("Smooth Image")
plt.subplot(1,5,5)
plt.imshow(sketch_image1, cmap='gray')
plt.title("Sketch Image")


plt.figure()
plt.subplot(1,2,1)
plt.imshow(sketch_image1, cmap='gray')
plt.title("Sketch Image - only gray")
plt.subplot(1,2,2)
plt.imshow(sketch_image2, cmap='gray')
plt.title("Sketch Image - inverted")
```
